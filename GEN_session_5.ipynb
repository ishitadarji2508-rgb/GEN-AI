2.Create a simple Q&A chatbot for Zomato-style restaurant info using any language or platform you know (Python, JavaScript, etc.): given a restaurant name and cuisine, return a canned response like 'Yes, Pizza Palace serves Italian cuisine.


```python
# Simple Zomato-Style Restaurant Chatbot (Python)

# Restaurant database
restaurants = {
    "Domino's": "Pizza",
    "McDonald's": "Fast Food",
    "Subway": "Sandwiches",
    "KFC": "Fried Chicken"
}


restaurant_name = input("Enter restaurant name: ")
cuisine = input("Enter cuisine: ")

# Check restaurant information
if restaurant_name in restaurants:
    if restaurants[restaurant_name].lower() == cuisine.lower():
        print(f"Yes, {restaurant_name} serves {cuisine} cuisine.")
    else:
        print(f"No, {restaurant_name} does not serve {cuisine} cuisine.")
else:
    print("Restaurant not found in the database.")

```

5.Use GitHub Copilot or ChatGPT to help you write a function that takes a list of song names and returns only those that start with the letter 'S', like a Spotify playlist filter.<br><br><em><strong>Constraint:</strong> You must show the prompt you used and the code Copilot or ChatGPT generated.</em>

```python
def filter_songs_starting_with_s(song_list):
    return [song for song in song_list if song.startswith('S')]

songs = ["Shape of You", "Believer", "Sanam Re ", "Thunder", "Senorita" , "Saiyaara"]
filtered_songs = filter_songs_starting_with_s(songs)

print(filtered_songs)
```
